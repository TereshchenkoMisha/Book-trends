```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:\\data_vis\\data\\raw\\goodreads_dataset_copy.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['literary_awards'] = df['literary_awards'].fillna('None') #if there is no awards, fill with 'None'
```

```python
df.describe()
```

```python
df['pages'] = df['pages'].fillna(df['pages'].median()) #fill the value in 'pages' with median
```

```python
df['language'] = df['language'].fillna('English') #if there are noo language value, put English
```

```python
df['format'] = df['format'].fillna(df['format'].mode()[0]) #fill format with mode value (the most popular)
```

```python
df['original_title'] = df['original_title'].fillna(df['title'])#if there are no title, fill with 'title'
```

```python
df['publication_date'] = df['publication_date'].fillna("Unknown") #if there are no publication date fill with 'Unknown'
```

```python
df['isbn'] = df['isbn'].fillna("Unknown") #if there are no isbn fill with 'Unknown'
```

```python
df['description'] = df['description'].fillna("") #put " " instead of empty description
```

```python
#replace Null values of ['series']
def clean_series(row):
    title = str(row['title'])
    current_series = str(row['series'])
    
    #try to retrieve the name of series from title
    if pd.notna(row['series']) and str(row['series']).strip() != '':
        return str(row['series']).strip()
    elif ':' in title:
        return title.split(':')[0].strip()
    elif '-' in title:
        return title.split('-')[0].strip()


df['series'] = df.apply(clean_series, axis=1)
```
